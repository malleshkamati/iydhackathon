```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import UnexpectedAlertPresentException, TimeoutException, NoSuchElementException
import pandas as pd
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup Chrome WebDriver
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # Uncomment for headless mode
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# URL of the website to scrape
url = "https://www.valmiki.iitk.ac.in/cbkv"
driver.get(url)

# Function to handle unexpected alerts
def handle_alert():
    try:
        WebDriverWait(driver, 2).until(EC.alert_is_present())  # Wait if alert exists
        alert = driver.switch_to.alert
        print(f"Unexpected Alert: {alert.text}")
        alert.accept()
        time.sleep(2)  # Allow page to stabilize
    except TimeoutException:
        pass  # No alert present, continue execution

# Wait for dropdowns to load
wait = WebDriverWait(driver, 10)
script_dropdown = wait.until(EC.presence_of_element_located((By.ID, "edit-language")))
sarga_dropdown = wait.until(EC.presence_of_element_located((By.ID, "edit-field-sarga-value")))
sloka_dropdown = wait.until(EC.presence_of_element_located((By.ID, "edit-field-sloka-value")))

# Convert to Select elements
script_dropdown = Select(script_dropdown)
sarga_dropdown = Select(sarga_dropdown)
sloka_dropdown = Select(sloka_dropdown)

# Data storage
data = []

# Loop through each script
for script_option in script_dropdown.options:
    script = script_option.text
    script_option.click()
    handle_alert()  # Handle alerts
    time.sleep(1)

    # Loop through each Sarga
    for sarga_option in sarga_dropdown.options:
        sarga = sarga_option.text
        sarga_option.click()
        handle_alert()  # Handle alerts
        time.sleep(1)

        # Loop through each Sloka
        for sloka_option in sloka_dropdown.options:
            sloka = sloka_option.text
            try:
                sloka_option.click()
                handle_alert()  # Handle alerts
                time.sleep(2)  # Wait for content to load
            except UnexpectedAlertPresentException:
                handle_alert()  # Accept the alert before retrying
                sloka_option.click()
                time.sleep(2)  # Let content load

            # Extract Sanskrit verse
            try:
                handle_alert()  # Handle alerts before finding element
                sanskrit_text_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "views-field-body")))
                sanskrit_text = sanskrit_text_element.text if sanskrit_text_element else "Not Found"
            except (TimeoutException, NoSuchElementException):
                sanskrit_text = "Not Found"

            # Extract Explanation
            try:
                handle_alert()  # Handle alerts before finding element
                explanation_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "views-field-field-explanation")))
                explanation = explanation_element.text if explanation_element else "Not Found"
            except (TimeoutException, NoSuchElementException):
                explanation = "Not Found"

            # Store the data
            data.append([script, sarga, sloka, sanskrit_text, explanation])
            print(f"Scraped: {script} - Sarga {sarga}, Sloka {sloka}")

# Save to CSV
df = pd.DataFrame(data, columns=["Script", "Sarga", "Sloka", "Sanskrit Text", "Explanation"])
df.to_csv("ramayana_slokas.csv", index=False, encoding="utf-8")
print("Scraping Completed! Data saved as ramayana_slokas.csv")

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import UnexpectedAlertPresentException, TimeoutException, NoSuchElementException
import pandas as pd
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup Chrome WebDriver
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # Uncomment for headless mode
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# URL of the website to scrape
url = "https://www.valmiki.iitk.ac.in/cbkv"
driver.get(url)

# Function to handle unexpected alerts
def handle_alert():
    try:
        WebDriverWait(driver, 2).until(EC.alert_is_present())  # Wait if alert exists
        alert = driver.switch_to.alert
        print(f"Unexpected Alert: {alert.text}")
        alert.accept()
        time.sleep(2)  # Allow page to stabilize
    except TimeoutException:
        pass  # No alert present, continue execution

# Wait for the dropdowns to load
wait = WebDriverWait(driver, 10)

# Data storage
data = []
sample_limit = 10  # Limit samples
sample_count = 0  # Counter

# Loop through each script
script_dropdown = Select(wait.until(EC.presence_of_element_located((By.ID, "edit-language"))))
for script_index in range(len(script_dropdown.options)):
    script_dropdown = Select(driver.find_element(By.ID, "edit-language"))  # Re-fetch
    script = script_dropdown.options[script_index].text
    script_dropdown.select_by_index(script_index)
    handle_alert()  # Handle alerts
    time.sleep(1)

    # Loop through each Sarga
    sarga_dropdown = Select(driver.find_element(By.ID, "edit-field-sarga-value"))  # Re-fetch
    for sarga_index in range(len(sarga_dropdown.options)):
        sarga_dropdown = Select(driver.find_element(By.ID, "edit-field-sarga-value"))  # Re-fetch
        sarga = sarga_dropdown.options[sarga_index].text
        sarga_dropdown.select_by_index(sarga_index)
        handle_alert()
        time.sleep(1)

        # Loop through each Sloka
        sloka_dropdown = Select(driver.find_element(By.ID, "edit-field-sloka-value"))  # Re-fetch
        for sloka_index in range(len(sloka_dropdown.options)):
            sloka_dropdown = Select(driver.find_element(By.ID, "edit-field-sloka-value"))  # Re-fetch
            sloka = sloka_dropdown.options[sloka_index].text
            try:
                sloka_dropdown.select_by_index(sloka_index)
                handle_alert()
                time.sleep(2)  # Wait for content to load
            except UnexpectedAlertPresentException:
                handle_alert()
                sloka_dropdown.select_by_index(sloka_index)
                time.sleep(2)

            # Extract Sanskrit verse
            try:
                handle_alert()
                sanskrit_text_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "views-field-body")))
                sanskrit_text = sanskrit_text_element.text if sanskrit_text_element else "Not Found"
            except (TimeoutException, NoSuchElementException):
                sanskrit_text = "Not Found"

            # Extract Explanation
            try:
                handle_alert()
                explanation_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "views-field-field-explanation")))
                explanation = explanation_element.text if explanation_element else "Not Found"
            except (TimeoutException, NoSuchElementException):
                explanation = "Not Found"

            # Store the data
            data.append([script, sarga, sloka, sanskrit_text, explanation])
            print(f"Scraped: {script} - Sarga {sarga}, Sloka {sloka}")

            # Stop after collecting 10 samples
            sample_count += 1
            if sample_count >= sample_limit:
                break
        if sample_count >= sample_limit:
            break
    if sample_count >= sample_limit:
        break

# Save to CSV
df = pd.DataFrame(data, columns=["Script", "Sarga", "Sloka", "Sanskrit Text", "Explanation"])
df.to_csv("ramayana_slokas_sample.csv", index=False, encoding="utf-8")
print("Scraping Completed! Data saved as ramayana_slokas_sample.csv")

# Close the browser
driver.quit()
```
